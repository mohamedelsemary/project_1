```python
#import requirement
from google.colab.patches import cv2_imshow
import cv2
import easyocr
```

```python
#use Reader model from easyocr for english
reader=easyocr.Reader(["en"])
```

```python
#read image and show it
def image_extract(image_path):
  imge=cv2.imread(image_path)
  #use reader model to predict my new image
  results=reader.readtext(imge)
  #show the character recognition in image
  for result in results :
    for val in result:
      to_lif=tuple(result[0][0])
      to_rit=tuple(result[0][2])
      text=result [1]
      text2=result[2]
      y=len(text)*20
      formatted_number = "{:.2f}".format(text2)
      #print(formatted_number)
      img1=cv2.rectangle(imge,(to_lif),(to_rit),(200,150,100) , 1 )
      img1 = cv2.putText(imge, text , to_lif, cv2.FONT_HERSHEY_SIMPLEX, 1, (100, 100, 255), 1)
      img1 = cv2.putText(imge, formatted_number , (to_lif[0]+y,to_lif[1]), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (100, 100, 255), 1)
  cv2_imshow(img1)
```

```python
image_extract("/content/1066.jpg")
```

```python
def extract_text_from_video(video_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    # Initialize the EasyOCR reader
    reader = easyocr.Reader(['en'])

    # Initialize an empty list to store the extracted text
    extracted_text = []

    while video.isOpened():
        # Read the current frame
        ret, frame = video.read()

        if not ret:
            break

        # Perform OCR on the frame
        results = reader.readtext(frame)

        # Extract the text and bounding boxes from the OCR result
        for result in results :
          for val in result:
             to_lif=tuple(result[0][0])
             to_rit=tuple(result[0][2])
             text=result [1]
             text2=result[2]
             y=len(text)*10
             formatted_number = "{:.2f}".format(text2)
             #print(formatted_number)
          img1=cv2.rectangle(frame,(to_lif),(to_rit),(0,0,0) , 1 )
          img1 = cv2.putText(frame, text , to_lif, cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 1)
          img1 = cv2.putText(frame, formatted_number , (to_lif[0]+y,to_lif[1]), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 1)

        # Display the frame
          cv2_imshow(img1)
          if cv2.waitKey(24) & 0xFF == ord('q'):

              video.release()
              cv2.destroyAllWindows()
              break


    # Release the video file
```
